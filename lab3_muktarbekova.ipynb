```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

```python
x = np.linspace(-10, 10, 100)  
y = 1 * x**2 -2 * x + 1
plt.figure(figsize=(8, 6)) 
plt.plot(x, y, label=f'y = {1}x^2 - {2}x + {1}') 
plt.xlabel('X-Axis') 
plt.ylabel('Y-Axis')
plt.title('График квадратичной функции') 
plt.grid(True)
plt.show()
```

* Построить график произвольной гармонической функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

```python
# Определяем параметры для гармонической функции
amplitude1 = 2.0
frequency1 = 1.0
phase1 = 0.0
amplitude2 = 1.5
frequency2 = 2.0
phase2 = np.pi/2

x = np.linspace(0, 2*np.pi, 100)
y1 = amplitude1 * np.sin(frequency1 * x + phase1)
y2 = amplitude2 * np.sin(frequency2 * x + phase2)
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y1 = 2*sin(x)', color='blue')
plt.plot(x, y2, label='y2 = 1.5*sin(2x)', color='red', linestyle='--')
plt.xlabel('X-Axis')  
plt.ylabel('Y-Axis') 
plt.title('График гармонических функций')   
plt.grid(True)  
plt.show()
```

* Вывести графики п.1 и 2 в разные поля при помощи функции subplot()

```python
#График квадратичной функции
a = 1
b = -2
c = 1
x1 = np.linspace(-10, 10, 100)
y1 = a * x1**2 + b * x1 + c

#График гармонических функций
amplitude1 = 2.0
frequency1 = 1.0
phase1 = 0.0
amplitude2 = 1.5
frequency2 = 2.0
phase2 = np.pi/2
x2 = np.linspace(0, 2*np.pi, 100)
y2_1 = amplitude1 * np.sin(frequency1 * x2 + phase1)
y2_2 = amplitude2 * np.sin(frequency2 * x2 + phase2)

plt.figure(figsize=(12, 6))  

# Подграфик 1 
plt.subplot(1, 2, 1)
plt.plot(x1, y1, label=f'y = {a}x^2 + {b}x + {c}', color='blue')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('График квадратичной функции')
plt.grid(True)

# Подграфик 2
plt.subplot(1, 2, 2)
plt.plot(x2, y2_1, label='y1 = 2*sin(x)', color='red')
plt.plot(x2, y2_2, label='y2 = 1.5*sin(2x)', color='green', linestyle='--')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('График гармонических функций')
plt.grid(True)

plt.tight_layout() 
plt.show()
```

* Построить графики п.1 и 2 в одной координатной плоскости, сделать подписи графиков, указать легенду.

```python
#График квадратичной функции
a = 1
b = -2
c = 1
x1 = np.linspace(-10, 10, 100)
y1 = a * x1**2 + b * x1 + c

#График гармонических функций
amplitude1 = 2.0
frequency1 = 1.0
phase1 = 0.0
amplitude2 = 1.5
frequency2 = 2.0
phase2 = np.pi/2
x2 = np.linspace(0, 2*np.pi, 100)
y2_1 = amplitude1 * np.sin(frequency1 * x2 + phase1)
y2_2 = amplitude2 * np.sin(frequency2 * x2 + phase2)

plt.figure(figsize=(10, 6))
plt.plot(x1, y1, label=f'y = {a}x^2 + {b}x + {c}', color='blue')
plt.plot(x2, y2_1, label='y1 = 2*sin(x)', color='red')
plt.plot(x2, y2_2, label='y2 = 1.5*sin(2x)', color='green', linestyle='--')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Графики квадратичной и гармонических функций')
plt.legend()

plt.grid(True)
plt.show()
```

* Построить график функции, заданной параметрически, x(t)=cos(3t),y=sin(2t),t∈[0,4π]

```python
t = np.linspace(0, 4 * np.pi, 1000)
x = np.cos(3 * t)
y = np.sin(2 * t)
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.title('График параметрической функции')
plt.plot(x, y)
```

* Построить столбиковую диаграмму по данным статистического комитета по прожиточному минимуму

```python
data_dates = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
data_values = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]
plt.figure(figsize=(12, 6))
plt.xlabel('Year')
plt.ylabel('Living wage')
plt.bar(data_dates, data_values)
plt.title('Диаграмма по данным статистического комитета КР по прожиточному минимуму')
plt.show()
```

* Построить круговой график по данным выше

```python
plt.figure(figsize=(10, 10))
plt.pie(data_values, labels=data_dates, autopct='%1.1ff%%', explode=[0.1] + [0 for _ in range(len(data_dates) - 1)])
plt.show()
```

* Построить график произвольной каркасной поверхности



```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z, cmap='viridis', linewidth=0.5)
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_zlabel('Z-Axis')
ax.set_title('Каркасная поверхность')
plt.show()
```
